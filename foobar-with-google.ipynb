```python
def solution (area):
    from math import sqrt
    if area < 1 or area > 1000000:
        raise ValueError("area should lie between 1 and 1000000 inclusive")
    if sqrt(area).is_integer():
        return[area]
    biggest_area = int(sqrt(area))**2
    return [biggest_area] + solution(area - biggest_area)
```

```python
solution(12)
```

```python
def solution (area):
    square_solar_panels = []
    while area > 0:
        largest_square_yard_side = int(area ** 0.5)
        largest_square_yard_area = largest_square_yard_side ** 2
        area -= largest_square_yard_area
        square_solar_panels.append(largest_square_yard_area)
        
    return square_solar_panels
```
